```python
import cv2
from glob import glob
```

```python
model = 'monocularDepth_'
# model = 'roadSegmentation_'

# img_folders = ['bag1','bag2']
img_folders = ['bag3','bag4']
```

```python
for img_folder in img_folders:
    
    with open(model+'predictions/'+img_folder+'/predictions.csv', 'r') as file:
        lines = file.readlines()

    lines_prediction = dict([line.strip().split(',') for line in lines])

    with open('groundTruth_'+img_folder+'.csv', 'r') as file:
        lines = file.readlines()

    lines_depthWayPoint = dict([line.strip().split(',') for line in lines])
    
    filenames = glob('orchard_images/'+img_folder+'/frame*.png')
    
    for filename in filenames:
        img = cv2.imread(filename)

        overlay = img.copy()
        output = img.copy()

        alpha = 0.5

        frame = filename.split('frame')[-1].split('.')[0]
        predictedPathMid = lines_prediction[frame]
        depthPathMid = lines_depthWayPoint[frame]

        waypoint = cv2.circle(overlay, (int(predictedPathMid),400), 150, (255,0,0), -1)
        waypoint = cv2.circle(overlay, (int(depthPathMid),400), 150, (0,255,0), -1)
        waypoint = cv2.addWeighted(overlay, alpha, output, 1 - alpha, 0, output)

        cv2.imwrite(model+'predictions/predicted_images/frame_'+img_folder+'_'+'{:03d}'.format( int(frame) )+'.png', output)
```
