```python
import cv2
from glob import glob
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
models = ['roadSegmentation_','monocularDepth_']

img_folders = ['bag1','bag2','bag3','bag4']
```

```python
for model in models:
    
    print ("\n-----\n"+model)
    
    for img_folder in img_folders:
        
#         if int(img_folder[-1])>2 and model=='roadSegmentation_':
#             continue

        print ("\n-----\n"+img_folder)

        with open(model+'predictions/'+img_folder+'/predictions.csv', 'r') as file:
            lines = file.readlines()

        predicted_lines = dict([(int(line.strip().split(',')[0]),line.strip().split(',')[1]) for line in lines])

        with open('groundTruth_'+img_folder+'.csv', 'r') as file:
            lines = file.readlines()

        depthWayPoints_lines = dict([(int(line.strip().split(',')[0]),line.strip().split(',')[1]) for line in lines])

        depthWaypointDeg = [ math.degrees(math.atan2(320-0, int(depthWayPoints_lines[i])-640)) for i in range(1,len(depthWayPoints_lines.keys())+1)]
        predictedDeg = [ math.degrees(math.atan2(320-0, int(predicted_lines[i])-640)) for i in range(1,len(depthWayPoints_lines.keys())+1)]
        
        predictionsIncorrect = np.where(abs(np.array(depthWaypointDeg)-np.array(predictedDeg))>25)
        print (predictionsIncorrect[0])
        # print (len(predictionsIncorrect[0]),len(depthWaypointDeg))
        print (len(predictionsIncorrect[0])/len(depthWaypointDeg))
```
